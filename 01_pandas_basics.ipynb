```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
avg_tip_by_day = df.groupby("day")["tip"].mean()
plt.figure(figsize=(5,4))
sns.barplot(x = avg_tip_by_day.index, y = avg_tip_by_day.values)
plt.title("average tip by day of weak")
plt.xlabel("days of weak")
plt.ylabel("average tip of day")
plt.show()
```

```python
avg_tip_by_time = df.groupby("time")["tip"].mean()
plt.figure(figsize=(5,5))
sns.barplot(x = avg_tip_by_time.index, y = avg_tip_by_time.values, hue = avg_tip_by_time.index, palette="Set2", legend = False)
plt.title("Average tip: Dinner Vs Lunch")
plt.xlabel("Meal Time")
plt.ylabel("average tip")
plt.show()
```

```python
avg_tip_by_sex = df.groupby("sex")["tip"].mean()
plt.figure(figsize=(5,5))
sns.barplot(x = avg_tip_by_sex.index, y= avg_tip_by_sex.values, hue = avg_tip_by_sex.index, palette="Set3", legend=False)
plt.title("Average Tip By Sex")
plt.xlabel("Gender")
plt.ylabel("Average Tip")
plt.show()
```

```python
avg_tip_by_smoker = df.groupby("smoker")["tip"].mean()
plt.figure(figsize=(5,5))
sns.barplot(x = avg_tip_by_smoker.index, y = avg_tip_by_smoker.values, hue = avg_tip_by_smoker.index, palette = "Set2", legend=False)
plt.title("Average Tip By Smoker")
plt.xlabel("smoker/non-smoker")
plt.ylabel("Average Tip")
plt.show()
```

```python
avg_tip_by_size = df.groupby("size")["tip"].mean()
plt.figure(figsize=(5,5))
sns.barplot(x = avg_tip_by_size.index, y = avg_tip_by_size.values, hue = avg_tip_by_size.index, palette = "Set3", legend=False)
plt.title("Average Tip By Group-Size")
plt.xlabel("Group Size")
plt.ylabel("Average Tip")
plt.show()
```
